```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
data=requests.get("https://www.flipkart.com/search?q=sunglasses&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&as-pos=2&as-type=RECENT&suggestionId=sunglasses%7CSunglasses&requestId=8ee38188-fccb-45a7-98dc-5d3053ab2a38&as-backfill=on")
```

```python
data
```

```python
soup=BeautifulSoup(data.content,'html.parser')
```

```python
s_brand=soup.find_all("div",class_='_2WkVRV')
s_brand
```

```python
brands=[]
for brand in s_brand:
    brands.append(brand.get_text("/n",""))
brands    
```
